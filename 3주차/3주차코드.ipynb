```python
import torch
import torch.nn as nn
```

```python
x = torch.full((3, 5), 1.0).requires_grad_()
```

```python
x
```

```python
x.size()
```

```python
dropout = nn.Dropout(p = 0.75)
```

```python
y = dropout(x)
```

```python
y
```

```python
l = y.norm(2, 1).sum()
```

```python
l.backward()
```

```python
x.grad
```

```python
x.grad.zero_()
```
